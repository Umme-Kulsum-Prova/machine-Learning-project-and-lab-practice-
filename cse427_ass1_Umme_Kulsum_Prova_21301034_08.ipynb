```python
import matplotlib.pyplot as plt

temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28,25,29,20,26,50,19,29,52,55]

study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]

subjects = ['Maths','English','Science','Physics','Computer']
subject_marks = [89,90,45,78,99]

#1
plt.figure()
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")
plt.show()

#2
plt.figure(figsize=(12,8))
plt.plot(study_hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()

#3
plt.figure()
plt.barh(subjects, subject_marks)
plt.title("Subject Scores (Horizontal)")
plt.show()

#4
plt.figure()
plt.bar(subjects, subject_marks, color=['red','blue','green','orange','purple'])
plt.title("Subject Scores (Vertical)")
plt.show()

#5
plt.figure(figsize=(14,10))

plt.subplot(2,2,1)
plt.scatter(temperature_dhaka, humidity_dhaka)
plt.title("Scatter")

plt.subplot(2,2,2)
plt.plot(study_hours, marks)
plt.title("Line Plot")

plt.subplot(2,2,3)
plt.barh(subjects, subject_marks)
plt.title("Horizontal Bar")

plt.subplot(2,2,4)
plt.bar(subjects, subject_marks)
plt.title("Vertical Bar")

plt.tight_layout()
plt.show()
```
